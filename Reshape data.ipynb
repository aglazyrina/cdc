```python
import pandas as pd
pd.options.display.max_columns = None
```

```python
df = pd.read_csv('svi_main_ep_spl.csv')
```

```python
df.head()
```

```python
#Check that every FIPS has one row
df.groupby('FIPS').count().sort_values(by  = 'AREA_SQMI_2014', ascending = False).head()
```

```python
#Reshaping the dataframe for easier manipulation in Tableau
dif_df = df.loc[:, df.columns.str.contains('2018_2014_DIF')]
#delete PCI because EP_PCI is the same as E_PCI
del dif_df['EP_PCI_2018_2014_DIF']
reshaped_df = pd.DataFrame()
for i in range(0,len(df)):
    print(i, len(df))
    reshaped_row = pd.DataFrame(dif_df.loc[i, dif_df.columns.str.contains('EP')].sort_values(ascending = False))
    reshaped_row = reshaped_row.reset_index()
    reshaped_row.columns  = ['factor', 'value']
    reshaped_row['factor'] = reshaped_row['factor'].apply(lambda x: x.split('_')[1])
    reshaped_row = reshaped_row.reset_index().rename(columns = {'index':'ranking'})
    reshaped_row['STATE'] = df.loc[i,'STATE']
    reshaped_row['COUNTY'] = df.loc[i,'COUNTY']
    reshaped_row['FIPS'] = df.loc[i,'FIPS']
    reshaped_df = reshaped_df.append(reshaped_row)
```

```python
reshaped_df.to_csv('reshaped_df.csv')
```
